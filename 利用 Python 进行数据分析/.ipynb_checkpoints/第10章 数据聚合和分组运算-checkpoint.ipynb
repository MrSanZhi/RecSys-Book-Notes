# 10.1 GroupBy机制

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one'],
                   'data1' : np.random.randn(5),
                   'data2' : np.random.randn(5)})
df
```

```python
# 按key1进行分组，并计算data1的平均值
# 访问data1,并根据key1调用groupby

grouped = df['data1'].groupby(df['key1'])
```

```python
# 变量grouped是一个GroupBy对象。
# 该对象拥有接下来对各分组执行运算所需的一切信息

grouped
```

```python
grouped.mean()
```

```python
# 一次传入多个数组的列表

means = df['data1'].groupby([df['key1'], df['key2']]).mean()
```

```python
means
```

```python
# 这里，通过两个键对数据进行了分组，得到的Sereis具有一个层次化索引（由唯一的键对组成）

means.unstack()
```

```python
# 在上例中，分组键均为Series。实际上，【分组键】可以是任何【长度适当】的数组

states = np.array(['Ohio', 'California', 'California', 'Ohio', 'Ohio'])
```

```python
years = np.array([2005, 2005, 2006, 2005, 2006])
```

```python
df['data1'].groupby([states, years]).mean()
```

```python
# 可以将列名（可以是字符串、数字或其他python对象）用作分组键
# 结果中没有key2列。因为df['key2']不是数值数据（俗称“麻烦列”），所以被从结果中排除了。

df.groupby('key1').mean()
```

```python
df.groupby(['key1', 'key2']).mean()
```

```python
# GroupBy的size方法：返回一个含有分组大小的Sereies
# 注意，任何分组关键词中的缺失值，都会被从结果中除去。

df.groupby(['key1', 'key2']).size()
```

## 10.1.1 对分组进行迭代

```python
import numpy as np
import pandas as pd
```

```python
# GroupBy对象支持迭代，可以产生一组【二元元组】（由分组名和数据块组成）

df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one'],
                   'data1' : np.random.randn(5),
                   'data2' : np.random.randn(5)})
df
```

```python
for name, group in df.groupby('key1'):
    print(name)
    print(group)
```

```python
# 对于多重键，元组的第一个元素将会是由键值组成的元组

for (k1, k2), group in df.groupby(['key1', 'key2']):
    print((k1, k2))
    print(group)
```

```python
# 将这些数据片段做成一个字典

pieces = dict(list(df.groupby('key1')))
pieces
```

```python
pieces['b']
```

```python
# 根据 dtype 对【列】进行分组

df.dtypes
```

```python
# groupby默认是在axis=0上进行分组。通过设置也可以在其他任何轴上进行分组

grouped = df.groupby(df.dtypes, axis=1)
```

```python
for dtype, group in grouped:
    print(dtype)
    print(group)
```

## 10.1.4 通过函数进行分组

```python
import numpy as np
import pandas as pd
```

```python
people = pd.DataFrame(np.random.randn(5, 5),
                       columns=['a', 'b', 'c', 'd', 'e'],
                       index=['Joe', 'Steve', 'Wes', 'Jim', 'Travis'])
people
```

## 10.2.1 面向列的多函数应用

```python
import numpy as np
import pandas as pd
```

```python
# 前面的小费例子。添加一个小费百分比的列tip_pct

tips = pd.read_csv('examples/tips.csv')
```

```python
# 加入小费占比

tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

```python
tips[ : 6]
```

```python
# 对不同的列使用不同的聚合函数，或一次应用多个函数
# 首先，根据 day 和 smoker 对 tips 进行分组

grouped = tips.groupby(['day', 'smoker'])
```

```python
# 注意，对于表10-1中那些描述统计，可以将函数名以字符串的形式传入

grouped_pct = grouped['tip_pct']
```

```python
# 将函数名以字符串的形式传入

grouped_pct.agg('mean')
```

```python
# 传入函数或函数名

grouped_pct.agg(['mean', 'std', peak_to_peak])
```

```python
# 如果要使用自定义的聚合函数，只需将其传入aggregate或agg方法即可

def peak_to_peak(arr):
    return arr.max() - arr.min()
```

```python
# 如果传入的是一个由(name, function)元组组成的列表，则各元组的第一个元素就会被用作DataFrame的列名
# 可以将这种二元元组列表看作一个有序映射

grouped_pct.agg([('foo', 'mean'), ('bar', np.std)])
```

```python
# 可以定义应用于全部列的函数，或不同的列应用不同的函数
# 对tip_pct和total_bill列计算三个统计信息

functions = ['count', 'mean', 'max']
```

```python
result = grouped['tip_pct', 'total_bill'].agg(functions)
result
```

```python
result['tip_pct']
```

```python
# 这里也可以传入带有自定义名称的一组【元组】

ftuples = [('Durchschnitt', 'mean'), ('Abweichung', np.var)]
```

```python
grouped['tip_pct', 'total_bill'].agg(ftuples)
```

```python
# 对一个列或不同的列应用不同的函数
# 方法：向agg传入一个列名映射到函数的【字典】
# 只有将多个函数应用到至少一列时，DataFrame才会拥有层次化的列

grouped.agg({'tip': np.max, 'size': 'sum'})
```

```python
grouped.agg({'tip_pct': ['min', 'max', 'mean', 'std'],
            'size': 'sum'})
```

# 10.3 apply: 一般性的“拆分-应用-合并”

```python
# 根据分组选出最高的5个tip_pct值
# 首先，编写一个选取指定列具有最大值的行的函数

def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n: ]
```

```python
top(tips, n=6)
```

```python
# 对smoker分组并用该函数调用apply

tips.groupby('smoker').apply(top)
```

```python
# 如果传给apply的函数能够接受其他参数或关键字，则可以将这些内容放在函数名后面一并传入

tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')
```

```python
result = tips.groupby('smoker')['tip_pct'].describe()
```

```python
result
```

```python
result.unstack('smoker')
```

## 10.3.2 分位数和桶分析

```python
frame = pd.DataFrame({'data1': np.random.randn(1000),
                       'data2': np.random.randn(1000)})
```

```python
quartiles = pd.cut(frame.data1, 4)
```

```python
quartiles[ : 10]
```

```python
# 由cut返回的Categorical对象可直接传递到groupby

def get_stats(group):
    return {'min': group.min(), 'max': group.max(),
           'count': group.count(), 'mean': group.mean()}
```

```python
grouped = frame.data2.groupby(quartiles)
```

```python
grouped.apply(get_stats).unstack()
```

```python
# 传入labels = False 即可只获取分位数的编号

grouping = pd.qcut(frame.data1, 10, labels=False)
```

```python
grouped = frame.data2.groupby(grouping)
```

```python
grouped.apply(get_stats).unstack()
```

## 10.3.3 示例：用特定于分组的值填充缺失值

```python
# 用平均值填充NA值

s = pd.Series(np.random.randn(6))
```

```python
s[ : : 2] = np.nan
```

```python
s
```

```python
s.fillna(s.mean())
```

```python
# 对不同的分组填充不同的值
# 将数据分组，并使用apply和一个能够对各数据块调用fillna的函数。

states = ['Ohio', 'New York', 'Vermont', 'Florida',
           'Oregon', 'Nevada', 'California', 'Idaho']
```

```python
group_key = ['East'] * 4 + ['West'] * 4
```

```python
data = pd.Series(np.random.randn(8), index=states)
```

```python
data
```

```python
# 将一些值设为缺失

data[['Vermont', 'Nevada', 'Idaho']] = np.nan
```

```python
data
```

```python
# 由group_key分组

data.groupby(group_key).mean()
```

```python
# 用分组平均值去填充NA值

fill_mean = lambda g: g.fillna(g.mean())
```

```python
data.groupby(group_key).apply(fill_mean)
```

```python
# 也可以在代码中预定义各组的填充值
# 利用分组的name属性

fill_values = {'East': 0.5, 'West': -1}
```

```python
fill_func = lambda g: g.fillna(fill_values[g.name])
```

```python
data.groupby(group_key).apply(fill_func)
```

## 10.3.4 示例：随机采样和排列

```python
suits = ['H', 'S', 'C', 'D']
card_val = (list(range(1, 11)) + [10] * 3) * 4
base_names = ['A'] + list(range(2, 11)) + ['J', 'K', 'Q']
cards = []
for suit in ['H', 'S', 'C', 'D']:
    cards.extend(str(num) + suit for num in base_names)

deck = pd.Series(card_val, index=cards)
```

```python
deck[ : 13]
```

```python
# 从整体中抽出5个

def draw(deck, n=5):
    return deck.sample(n)
```

```python
draw(deck)
```

```python
# 假设想要从每种花色中随机抽取两张牌。
# 由于花色是牌名的最后一个字符，所以可以据此进行分组，并使用apply

get_suit = lambda card: card[-1]
```

```python
deck.groupby(get_suit).apply(draw, n=2)
```

```python
# 也可以这样写

deck.groupby(get_suit, group_keys=False).apply(draw, n=2)
```

## 10.3.5 示例：分组加权平均数和相关系数

```python
# 根据groupby的“拆分-应用-合并”范式，可以进行DataFrame的列与列之间或两个Series之间的运算（比如分组加权平均）

df = pd.DataFrame({'category': ['a', 'a', 'a', 'a',
                                 'b', 'b', 'b', 'b'],
                    'data': np.random.randn(8),
                    'weights': np.random.rand(8)})
```

```python
grouped.apply(get_wavg)
```

```python
# 例二：考虑一个来自Yahoo!Finance的数据集

close_px = pd.read_csv('examples/stock_px_2.csv', parse_dates=True,
                      index_col=0)
```

```python
close_px.info()
```

```python
close_px[-4: ]
```

```python
# 计算一个由日收益率（通过百分数变化计算）与SPX之间的制度相关系数组成的DataFrame

spx_corr = lambda x: x.corrwith(x['SPX'])
```

```python
# 接下来使用pct_change计算close_px的百分比变化

rets = close_px.pct_change().dropna()
```

```python
# 最后，我们用【年】对百分比变化进行分组，可以用一个一行的聚光灯，
# 从每行的标签返回每个datetime标签的year属性

get_year = lambda x: x.year
```

```python
by_year = rets.groupby(get_year)
```

```python
by_year.apply(spx_corr)
```

```python
# 计算列与列之间的相关系数
# 这里，计算Apple和Microsoft的年相关系数

by_year.apply(lambda g: g['AAPL'].corr(g['MSFT']))
```

## 10.3.6 示例：组级别的线性回归

```python
# 可以用groupby执行更为复杂的分组统计分析，只要函数返回的是pandas对象或标量值即可
# 例如：可定义一个regress函数（利用statsmodels计量经济学库）对各数据执行普通最小二乘法回归

import statsmodels.api as sm

def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept'] = 1.
    result = sm.OLS(Y, X).fit()
    return result.params
```

```python
# 现在，为了按年计算AAPL对SPX收益率的线性回归执行：

by_year.apply(regress, 'AAPL', ['SPX'])
```

# 10.4 透视表和交叉表

```python
# pivot table： 数据透视表
# DataFrame_instance.pivot_table() or pandas.pivot_table()
# 回到小费数据集。根据day和smoker计算分组平均数（pivot_table的默认聚合类型）,并将day和smoker放到行上

s = pd.read_csv('examples/tips.csv')

s.pivot_table(index=['day', 'smoker'])
```

```python
# 如果只想聚合tip_pct和size,而且想根据time进行分组。
# 将smoker放到列上，把day放到行上

tips.pivot_table(['tip', 'size'], index=['time', 'day'],
                columns='smoker')
```

```python
# 传入margins=True添加分项小计.
# 这将会添加【标签为All】的行和列，其值对应于单个等级中所有数据的分组统计

tips.pivot_table(['tip', 'size'], index=['time', 'day'],
                columns='smoker', margins=True)

# 这里，All值为平均数：不单独考虑烟民与非烟民（All列），不单独考虑行分组两个级别中的任何单项（All列）
```

```python
# 要使用其它的聚合函数，将其传给aggfunc即可
# 如：使用count或len可以得到有关分组大小的交叉表(计数或频率)

tips.pivot_table('tip', index=['time', 'smoker'],
                aggfunc=len, margins=True)
```

```python
# 如果存在空的组合（也就是NA），设置一个fill_value

tips.pivot_table('tip', index=['time', 'size', 'smoker'],
                columns='day', aggfunc='mean', fill_value=0)
```

## 10.4.1 交叉表：crosstab

```python
# 交叉表（cross-tabulation, crosstab）是一种用于计算【分组频率】的特殊透视表

from io import StringIO

data = """\
Sample  Nationality  Handedness
1   USA  Right-handed
2   Japan    Left-handed
3   USA  Right-handed
4   Japan    Right-handed
5   Japan    Left-handed
6   Japan    Right-handed
7   USA  Right-handed
8   USA  Left-handed
9   Japan    Right-handed
10  USA  Right-handed"""

data = pd.read_csv(StringIO(data), sep='\s+')
```

```python
# 根据国籍和用手习惯对这段进行统计汇总

pd.crosstab(data.Nationality, data.Handedness, margins=True)
```

```python
# crosstab的前两个参数可以是数组或Series，或是数组列表。
# 就像小费数据

pd.crosstab([tips.time, tips.day], tips.smoker, margins=True)
```
